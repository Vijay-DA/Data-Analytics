```python
### Task 1:

"""Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line. """

# Initialize an empty list to store the numbers
result_numbers = []

# Iterate through numbers from 2000 to 3200 (inclusive)
for i in range(2000, 3201):
    # Check if the number is divisible by 7 and not a multiple of 5
    if (i % 7 == 0) and (i % 5 != 0):
        result_numbers.append(str(i))

# Join the numbers with a comma and print them
print(",".join(result_numbers))
```

```python
"""Write a Python program to accept the user's first and last name and then getting them printed in
the the reverse order with a space between first name and last name."""


# Accept user's first name
first_name = input("Enter your first name: ")

# Accept user's last name
last_name = input("Enter your last name: ")

# Print them in reverse order with a space in between
print(f"{last_name} {first_name}")
```

```python
""" Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 * π * r """

import math

# Given diameter
diameter = 12 # cm

# Calculate radius
radius = diameter / 2

# Calculate volume using the formula V = 4/3 * pi * r^3
volume = (4/3) * math.pi * (radius ** 3)

print(f"The diameter of the sphere is: {diameter} cm")
print(f"The radius of the sphere is: {radius} cm")
print(f"The volume of the sphere is: {volume:.2f} cubic cm")
```

```python
## Task 2:

"""Write a program which accepts a sequence of comma-separated numbers from console and
generate a list."""

# Accept a sequence of comma-separated numbers from the console
numbers_str = input("Enter comma-separated numbers: ")

# Generate a list by splitting the string at each comma
numbers_list = numbers_str.split(',')

# Print the generated list
print(f"The generated list is: {numbers_list}")
```

```python
"""Create the below pattern using nested for loop in Python.

*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

"""


# Loop for the increasing part of the pattern
for i in range(1, 6):
    print("* " * i)

# Loop for the decreasing part of the pattern
for i in range(4, 0, -1):
    print("* " * i)
```

```python
"""Write a Python program to reverse a word after accepting the input from the user.
Sample Output:
Input word: AcadGild
Output: dilGdacA """

# Accept input word from the user
input_word = input("Input word: ")

# Reverse the word using slicing
reversed_word = input_word[::-1]

# Print the reversed word
print(f"Output: {reversed_word}")
```
