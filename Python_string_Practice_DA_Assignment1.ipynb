```python
#Python String Practice Exercises
#(Data Analysis Focus)

"""LEVEL 1: Basics – Data Cleaning
Exercise 1: Remove extra spaces
Input:
name = " john doe "
Task: Remove leading and trailing spaces.
Expected Output:
"john doe"""

name = " john doe "
cleaned_name = name.strip()
print(f"Original: '{name}'")
print(f"Cleaned: '{cleaned_name}'")
```

```python
"""Exercise 2: Standardize case
Input:
city = "MuMbAi"
Task: Convert text to lowercase.
Expected Output:
"mumbai" """
city = "MuMbAi"
lowercase_city = city.lower()
print(f"Original: '{city}'")
print(f"Lowercase: '{lowercase_city}'")

```

```python
"""Exercise 3: Capitalize names
Input:
employee = "rahul sharma"
Task: Capitalize the first letter of each word.
Expected Output:
"Rahul Sharma" """

employee = "rahul sharma"
capitalized_employee = employee.title()
print(f"Original: '{employee}'")
print(f"Capitalized: '{capitalized_employee}'")
```

```python
"""Exercise 4: Replace missing values
Input:
status = "N/A"
Task: Replace "N/A" with "Not Available". """

status = "N/A"
updated_status = status.replace("N/A", "Not Available")
print(f"Original: '{status}'")
print(f"Updated: '{updated_status}'")
```

```python
"""LEVEL 2: Splitting and Parsing
Exercise 5: Split a CSV row
Input:
row = "101,Rahul,IT,45000"
Task: Convert the string into a list.
Expected Output:
["101", "Rahul", "IT", "45000"] """

row = "101,Rahul,IT,45000"
split_row = row.split(',')
print(f"Original: '{row}'")
print(f"Split: {split_row}")
```

```python
"""Exercise 6: Extract email domain
Input:
email = "user123@gmail.com"
Task: Extract the domain name.
Expected Output:
"gmail.com" """

email = "user123@gmail.com"
domain = email.split('@')[1]
print(f"Original Email: '{email}'")
print(f"Extracted Domain: '{domain}'")
```

```python
"""Exercise 7: Join values for reporting
Input:
skills = ["Python", "SQL", "Excel"]
Task: Join the list using | as a separator.
Expected Output:
"Python | SQL | Excel" """

skills = ["Python", "SQL", "Excel"]
joined_skills = " | ".join(skills)
print(f"Original List: {skills}")
print(f"Joined String: '{joined_skills}'")
```

```python
"""LEVEL 3: Validation – Data Quality Checks
Exercise 8: Validate numeric value
Input:
value = "45000"
Task: Check whether the value contains only digits.
Expected Output:
True """

value = "45000"
is_digit = value.isdigit()
print(f"Original value: '{value}'")
print(f"Contains only digits: {is_digit}")
```

```python
"""Exercise 9: Validate name field
Input:
name = "Ravi123"
Task: Check if the name contains only alphabets.
Expected Output:
False"""

name = "Ravi123"
is_alpha = name.isalpha()
print(f"Original name: '{name}'")
print(f"Contains only alphabets: {is_alpha}")
```

```python
"""Exercise 10: Detect blank values
Input:
comment = " "
Task: Identify if the value is invalid (blank).
Expected Output:
"Invalid" """

comment = " "
is_blank = comment.strip() == ""

if is_blank:
    print("Invalid")
else:
    print("Valid")
```

```python
"""LEVEL 4: Searching and Analysis
Exercise 11: Count keyword occurrence
Input:
feedback = "good service good support good response"
Task: Count how many times "good" appears. """


feedback = "good service good support good response"
keyword = "good"
occurrences = feedback.count(keyword)
print(f"The keyword '{keyword}' appears {occurrences} times.")
```

```python
"""Exercise 12: Find error position
Input:
log = "INFO:Success ERROR:Timeout"
Task: Find the index position of "ERROR". """

log = "INFO:Success ERROR:Timeout"
keyword_to_find = "ERROR"
error_position = log.find(keyword_to_find)
print(f"The keyword '{keyword_to_find}' is found at index: {error_position}")
```

```python
"""LEVEL 5: Real Data Analyst Scenarios
Exercise 13: Normalize phone numbers
Input:
phone = "+91-98765-43210"
Task: Remove special characters.
Expected Output: "919876543210" """


phone = "+91-98765-43210"

# Remove special characters
cleaned_phone = phone.replace("-", "").replace("+", "")

print(f"Original phone number: '{phone}'")
print(f"Normalized phone number: '{cleaned_phone}'")
```

```python
"""Exercise 14: Extract year from date
Input:
date = "2025-01-23"
Task: Extract the year.
Expected Output:
"2025" """


date_str = "2025-01-23"
year = date_str.split('-')[0]
print(f"Original date: '{date_str}'")
print(f"Extracted year: '{year}'")
```

```python
"""Exercise 15: Mask sensitive information
Input:
card = "1234-5678-9876-5432"
Task: Mask all digits except the last four.
Expected Output:
"XXXX-XXXX-XXXX-5432" """


card = "1234-5678-9876-5432"

last_four_digits = card[-4:]

# Mask all digits except the last four, preserving dashes
masked_card = ''
for i in range(len(card) - 4):
    if card[i].isdigit():
        masked_card += 'X'
    else:
        masked_card += card[i]

masked_card += last_four_digits

print(f"Original card: '{card}'")
print(f"Masked card: '{masked_card}'")
```

```python
"""CHALLENGE EXERCISES
Exercise 16: Clean messy text column
Input:
data = " PYthon , sql , EXCEL "
Task: Convert to lowercase, remove spaces, and format as:
"python,sql,excel" """


data = " PYthon , sql , EXCEL "
cleaned_data = data.lower().replace(' ', '')

print(f"Original data: '{data}'")
print(f"Cleaned data: '{cleaned_data}'")
```

```python
"""Exercise 17: Case-insensitive replacement
Input:
text = "I like PYTHON and Python"
Task: Replace all forms of "python" with "java". """


text = "I like PYTHON and Python"

# Convert to lowercase for case-insensitive replacement
processed_text = text.lower()

# Replace all occurrences of 'python' with 'java'
updated_text = processed_text.replace("python", "java")

print(f"Original text: '{text}'")
print(f"Updated text: '{updated_text}'")
```

```python
"""Exercise 18: Word frequency count
Input:
text = "data science data analysis data"

Task: Count frequency of each word.
Expected Output:
{'data': 3, 'science': 1, 'analysis': 1} """


from collections import Counter

text = "data science data analysis data"

# Convert to lowercase and split into words
words = text.lower().split()

# Count the frequency of each word
word_frequency = Counter(words)

print(f"Original text: '{text}'")
print(f"Word frequency: {word_frequency}")
```
