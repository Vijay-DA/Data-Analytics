```python
"""
You are hired as a Data Analyst in a hospital chain.
Management wants to analyze:
1. Department-wise patient revenue
2. Average treatment cost
3. Identify high-risk patients (cost > 50,000)
4. Calculate patient stay duration
5. Monthly admission trends
6. Top performing department
You are given patient admission data.

Dataset (Save as: hospital_data.csv)
PatientID,Name,Department,AdmissionDate,DischargeDate,TreatmentCost,InsuranceT
ype
P101,Rahul,Cardiology,2023-01-05,2023-01-10,60000,Private
P102,Amit,Orthopedics,2023-02-12,2023-02-18,35000,Government
P103,Neha,Neurology,2023-03-01,2023-03-12,75000,Private
P104,Priya,Cardiology,2023-04-08,2023-04-14,50000,Private
P105,Karan,Orthopedics,2023-05-10,2023-05-16,40000,Government
P106,Simran,Neurology,2023-06-15,2023-06-25,90000,Private

Tasks to Perform
1. Load and inspect dataset
2. Convert dates to datetime
3. Calculate hospital stay duration
4. Identify high-cost patients
5. Department-wise total revenue
6. Average treatment cost
7. Monthly admission trend
8. Visualize department revenue

"""
```

```python
#1. Load and inspect dataset

import pandas as pd

# Dataset content from the problem description
data = """
PatientID,Name,Department,AdmissionDate,DischargeDate,TreatmentCost,InsuranceType
P101,Rahul,Cardiology,2023-01-05,2023-01-10,60000,Private
P102,Amit,Orthopedics,2023-02-12,2023-02-18,35000,Government
P103,Neha,Neurology,2023-03-01,2023-03-12,75000,Private
P104,Priya,Cardiology,2023-04-08,2023-04-14,50000,Private
P105,Karan,Orthopedics,2023-05-10,2023-05-16,40000,Government
P106,Simran,Neurology,2023-06-15,2023-06-25,90000,Private
"""

# Save the dataset to a CSV file named 'hospital_data.csv'
with open('hospital_data.csv', 'w') as f:
    f.write(data.strip())

# Load the dataset into a pandas DataFrame
df = pd.read_csv('hospital_data.csv')

print("Dataset loaded successfully. Displaying the first 5 rows:")
display(df.head())
```

```python
# 2. Convert dates to datetime

df['AdmissionDate'] = pd.to_datetime(df['AdmissionDate'])
df['DischargeDate'] = pd.to_datetime(df['DischargeDate'])

print("Dates converted to datetime format. Displaying updated DataFrame info:")
df.info()
```

```python
# 3. Calculate hospital stay duration
df['StayDuration'] = (df['DischargeDate'] - df['AdmissionDate']).dt.days

print("DataFrame with calculated StayDuration:")
display(df.head())
```

```python
# 4. Identify high-cost patients

high_cost_patients = df[df['TreatmentCost'] > 50000]
print("\nHigh-cost patients (TreatmentCost > 50,000):")
display(high_cost_patients)
```

```python
# 5. Department-wise total revenue

department_revenue = df.groupby('Department')['TreatmentCost'].sum().reset_index()
department_revenue = department_revenue.rename(columns={'TreatmentCost': 'TotalRevenue'})
print("\nDepartment-wise Total Revenue:")
display(department_revenue)
```

```python
# 6. Average treatment cost

average_treatment_cost = df['TreatmentCost'].mean()
print(f"\nAverage Treatment Cost: {average_treatment_cost:.2f}")
```

```python
# 7. Monthly admission trend

df['AdmissionMonth'] = df['AdmissionDate'].dt.to_period('M')
monthly_admissions = df['AdmissionMonth'].value_counts().sort_index().reset_index()
monthly_admissions.columns = ['Month', 'Admissions']

print("Monthly admission trends:")
display(monthly_admissions)
```

```python
# 8. Visualize department revenue

import matplotlib.pyplot as plt
import seaborn as sns

# Ensure department_revenue is available. If it was not run previously, it would need to be re-calculated.
department_revenue = df.groupby('Department')['TreatmentCost'].sum().reset_index()
department_revenue = department_revenue.rename(columns={'TreatmentCost': 'TotalRevenue'})

plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='TotalRevenue', data=department_revenue)
plt.title('Department-wise Total Revenue')
plt.xlabel('Department')
plt.ylabel('Total Revenue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
