```python
#wap where use all types of operator using var1 is 5 and var2 is 2
# Define variables
var1 = 5
var2 = 2

print(f"var1 = {var1}")
print(f"var2 = {var2}\n")

# --- Arithmetic Operators ---
print("--- Arithmetic Operators ---")
print("var1 + var2") #addition
print("var1 - var2") #subtraction
print("var1 * var2") #multiplication
print("var1 / var2") #division
print("var1 % var2") #modulus
print("var1 ** var2") #exponentiation
print("var1 // var2\n") #floor division

# --- Comparison Operators ---
print("--- Comparison Operators ---")
print(f"Equal (var1 == var2): {var1 == var2}")
print(f"Not Equal (var1 != var2): {var1 != var2}")
print(f"Greater Than (var1 > var2): {var1 > var2}")
print(f"Less Than (var1 < var2): {var1 < var2}")
print(f"Greater Than or Equal (var1 >= var2): {var1 >= var2}")
print(f"Less Than or Equal (var1 <= var2): {var1 <= var2}\n")

# --- Assignment Operators ---
print("--- Assignment Operators ---")
x = var1
print(f"x = var1: {x}")
x += var2 # x = x + var2
print(f"x += var2: {x}")
x -= var2 # x = x - var2
print(f"x -= var2: {x}")
x *= var2 # x = x * var2
print(f"x *= var2: {x}")
x /= var2 # x = x / var2
print(f"x /= var2: {x}\n")

# --- Logical Operators ---
print("--- Logical Operators ---")
print(f"var1 > 0 and var2 < 10: {var1 > 0 and var2 < 10}")
print(f"var1 > 10 or var2 < 0: {var1 > 10 or var2 < 0}")
print(f"not(var1 > 10): {not(var1 > 10)}\n")

# --- Identity Operators ---
print("--- Identity Operators ---")
list1 = [1, 2]
list2 = [1, 2]
list3 = list1
print(f"list1 is list3: {list1 is list3}") # Same object
print(f"list1 is list2: {list1 is list2}") # Different objects, same content
print(f"list1 is not list2: {list1 is not list2}\n")

# --- Membership Operators ---
print("--- Membership Operators ---")
my_list = [1, 3, 5, 7]
print(f"var1 in my_list: {var1 in my_list}")
print(f"var2 not in my_list: {var2 not in my_list}\n")

# --- Bitwise Operators ---
print("--- Bitwise Operators ---")
# Binary representation: var1 = 5 (0101), var2 = 2 (0010)
print(f"Bitwise AND (var1 & var2): {var1 & var2}") # 0000 (0)
print(f"Bitwise OR (var1 | var2): {var1 | var2}") # 0111 (7)
print(f"Bitwise XOR (var1 ^ var2): {var1 ^ var2}") # 0111 (7)
print(f"Bitwise NOT (~var1): {~var1}") # -(5+1) = -6
print(f"Left Shift (var1 << var2): {var1 << var2}") # 5 * (2**2) = 20 (0101 -> 10100)
print(f"Right Shift (var1 >> var2): {var1 >> var2}") # 5 // (2**2) = 1 (0101 -> 0001)
```

```python
"""Exercise 15: Mask sensitive information
Input:
card = "1234-5678-9876-5432"
Task: Mask all digits except the last four.
Expected Output:
"XXXX-XXXX-XXXX-5432" """


card = "1234-5678-9876-5432"

last_four_digits = card[-4:]

# Mask all digits except the last four, preserving dashes
masked_card = ''
for i in range(len(card) - 4):
    if card[i].isdigit():
        masked_card += 'X'
    else:
        masked_card += card[i]

masked_card += last_four_digits

print(f"Original card: '{card}'")
print(f"Masked card: '{masked_card}'")
```

```python
### Task 1:

"""Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line. """

# Initialize an empty list to store the numbers
result_numbers = []

# Iterate through numbers from 2000 to 3200 (inclusive)
for i in range(2000, 3201):
    # Check if the number is divisible by 7 and not a multiple of 5
    if (i % 7 == 0) and (i % 5 != 0):
        result_numbers.append(str(i))

# Join the numbers with a comma and print them
print(",".join(result_numbers))
```

```python
""" Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 * π * r """

import math

# Given diameter
diameter = 12 # cm

# Calculate radius
radius = diameter / 2

# Calculate volume using the formula V = 4/3 * pi * r^3
volume = (4/3) * math.pi * (radius ** 3)

print(f"The diameter of the sphere is: {diameter} cm")
print(f"The radius of the sphere is: {radius} cm")
print(f"The volume of the sphere is: {volume:.2f} cubic cm")
```
