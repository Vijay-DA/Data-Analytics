```python
"""
PYTHON ASSIGNMENT-1
# Customer purchase dataset
customers = [
{"id": 101, "name": "Amit", "products": ["Laptop", "Mouse", "Keyboard"], "amount": 75000},
{"id": 102, "name": "Priya", "products": ["Mobile", "Earphones"], "amount": 30000},
{"id": 103, "name": "Rahul", "products": ["Laptop", "Mobile"], "amount": 95000},
{"id": 104, "name": "Sneha", "products": ["Tablet", "Keyboard"], "amount": 40000},
{"id": 105, "name": "Karan", "products": ["Laptop", "Mouse"], "amount": 70000}
]
• Each dictionary represents one customer.
• Keys:
o "id" → Customer ID
o "name" → Customer name
o "products" → List of products they purchased
o "amount" → Total amount spent

This dataset is small but realistic enough for beginner-level analysis. It lets students practice with lists,
sets, dictionaries, and tuples while simulating a real e-commerce case study.
What you have to perform
1. Lists
• Extract all customer names into a list.
• Collect all products purchased (with duplicates).
• Find the top spender using a simple loop.
2. Sets
• Get unique products purchased.
• Identify products bought by multiple customers.
• Find all customers who purchased a Laptop.
3. Dictionaries
• Map customer name → amount spent.
• Map product → number of buyers.
• Calculate average spending across all customers.
4. Tuples
• Store each customer’s (name, amount) as tuples.
• Sort tuples by amount spent.
• Create a tuple of highest and lowest spenders.

"""
```

```python
customers = [
    {"id": 101, "name": "Amit", "products": ["Laptop", "Mouse", "Keyboard"], "amount": 75000},
    {"id": 102, "name": "Priya", "products": ["Mobile", "Earphones"], "amount": 30000},
    {"id": 103, "name": "Rahul", "products": ["Laptop", "Mobile"], "amount": 95000},
    {"id": 104, "name": "Sneha", "products": ["Tablet", "Keyboard"], "amount": 40000},
    {"id": 105, "name": "Karan", "products": ["Laptop", "Mouse"], "amount": 70000}
]
```

## 1. Lists

### 1.1. Extract all customer names into a list.

```python
customer_names = [customer['name'] for customer in customers]
print("Customer Names:", customer_names)
```

### 1.2. Collect all products purchased (with duplicates).

```python
all_products_purchased = []
for customer in customers:
    all_products_purchased.extend(customer['products'])
print("All Products Purchased (with duplicates):", all_products_purchased)
```

### 1.3. Find the top spender using a simple loop.

```python
top_spender = None
max_amount = -1

for customer in customers:
    if customer['amount'] > max_amount:
        max_amount = customer['amount']
        top_spender = customer

print("Top Spender:", top_spender['name'], "with amount:", top_spender['amount'])
```

## 2. Sets

### 2.1. Get unique products purchased.

```python
unique_products = set(all_products_purchased)
print("Unique Products Purchased:", unique_products)
```

### 2.2. Identify products bought by multiple customers.

```python
from collections import defaultdict

product_purchase_counts = defaultdict(int)
for customer in customers:
    for product in customer['products']:
        product_purchase_counts[product] += 1

products_bought_by_multiple_customers = [product for product, count in product_purchase_counts.items() if count > 1]
print("Products bought by multiple customers:", products_bought_by_multiple_customers)
```

### 2.3. Find all customers who purchased a Laptop.

```python
customers_who_bought_laptop = [customer['name'] for customer in customers if 'Laptop' in customer['products']]
print("Customers who purchased a Laptop:", customers_who_bought_laptop)
```

## 3. Dictionaries

### 3.1. Map customer name → amount spent.

```python
customer_name_to_amount = {customer['name']: customer['amount'] for customer in customers}
print("Customer Name to Amount Spent:", customer_name_to_amount)
```

### 3.2. Map product → number of buyers.

```python
product_to_num_buyers = defaultdict(int)
for customer in customers:
    for product in customer['products']:
        product_to_num_buyers[product] += 1

print("Product to Number of Buyers:", dict(product_to_num_buyers))
```

### 3.3. Calculate average spending across all customers.

```python
total_spending = sum(customer['amount'] for customer in customers)
average_spending = total_spending / len(customers)
print(f"Average spending across all customers: {average_spending:.2f}")
```

## 4. Tuples

### 4.1. Store each customer’s (name, amount) as tuples.

```python
customer_name_amount_tuples = [(customer['name'], customer['amount']) for customer in customers]
print("Customer (Name, Amount) Tuples:", customer_name_amount_tuples)
```

### 4.2. Sort tuples by amount spent.

```python
sorted_customer_tuples_by_amount = sorted(customer_name_amount_tuples, key=lambda x: x[1], reverse=True)
print("Customer Tuples sorted by Amount (descending):
", sorted_customer_tuples_by_amount)
```

### 4.3. Create a tuple of highest and lowest spenders.

```python
highest_spender_tuple = sorted_customer_tuples_by_amount[0]
lowest_spender_tuple = sorted_customer_tuples_by_amount[-1]

highest_and_lowest_spenders = (highest_spender_tuple, lowest_spender_tuple)
print("Tuple of (Highest Spender, Lowest Spender):", highest_and_lowest_spenders)
```
